### We will take into account only those book who have been rated by atleast 50 user

```python
# If you want to do this by earlier method then before merging ""ratings=ratings[ratings['user_id'].isin(y)]""
# and books dataframe apply conditon on ISBN also as z=ratings['ISBN'].value_counts(), z=z[z>=50], z.index
# ratings=ratings[ratings['ISBN'].isin(z)]
# ratings.merge(books,on="ISBN")
number_rating=ratings_with_books.groupby('title')['rating'].count().reset_index()  
number_rating
```

```python
number_rating=number_rating.rename(columns={'rating':'number_of_ratings'})
number_rating
```

### Now we have to merge ot into the ratings_with_books Dataframe

```python
ratings_with_books.head(2)
```

```python
final_df=ratings_with_books.merge(number_rating,on='title')
```

```python
final_df
```

```python
ratings_with_books.shape,final_df.shape
```

##### We will take into account only those book who have been rated by atleast 50 user

```python
final_df=final_df[final_df['number_of_ratings']>=50]
final_df
```

```python
#### If a same user have rated a book multiple times this can be bad for our model so we will clear the duplicates.
final_df=final_df.drop_duplicates(['title','user_id'])
final_df
```

```python
final_df.nunique()# In ratings_with_books user_id: 899 and in final_df it decrease to user_id : 888
# As that was those users which rated on books , WHO WERE RATED BY LESS THAN 50 USERS
```

```python
book_pivot=final_df.pivot_table(values='rating',
    index='title',
    columns='user_id')
book_pivot
```

##### fill NaN value 

```python
book_pivot.fillna(0,inplace=True)
book_pivot
```

```python
# As you can see this is a very sparse type Mostly filled with zeros.
# We don,t want to calculate distance from zeros as it will increse our computational load.
```

```python
#So we will convert it into sparse_matrix  so that internally distance from zeros will not be calculated
from scipy.sparse import csr_matrix
book_sparse=csr_matrix(book_pivot)
```

```python
list(book_sparse)
```

```python
#Will use KNN(classification and regression) brother Nearest Neighbour Algorithm that do Clustering.
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
model=NearestNeighbors(algorithm='brute')#n_neighbors=5 hyperparameter
#algorithm='auto'  'auto' will attempt to decide the most appropriate algorithm   hyperparameter
#algorithm='brute' We instruct the algo to calculate distance of all the points and then decide neighbors  
```

```python
model.fit(book_sparse)
```

```python
book_pivot.iloc[0,:].values.reshape(1,-1)
```

```python
distances,suggestion=model.kneighbors(book_pivot.iloc[237,:].values.reshape(1, -1),n_neighbors=6)
#reshaping is must for our model to understand the data
#Neighbors=6 because 1 book will be the 237 book itself   AGAIN IT IS HYPERPARAMETER
```

```python
def suggest(book_name):
    book_index=np.where(book_pivot.index==book_name)[0][0]
    distances,suggestion=model.kneighbors(book_pivot.iloc[book_index,:].values.reshape(1, -1),n_neighbors=6)
    for i in range(len(suggestion)):
        return 'The suggestion for the books: '+ str(book_pivot.index[suggestion[0][0]])+' are',book_pivot.index[suggestion[i]].values[1:6]
```
